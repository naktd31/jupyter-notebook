```python
import numpy as np
from scipy.optimize import minimize
from sklearn import linear_model
```

```python
n_s = 15
D = 10
d = 3
```

```python
x_t = np.zeros((D,1))
index = np.random.choice([i for i in range(D)], d, False)
x_t[index] = 1
#x_t = np.random.uniform(0, 1, D)
x_t.sum(), x_t
```

```python
A = np.zeros([D, n_s])

for j in range(n_s):
    index = np.random.choice([i for i in range(D)], d, False)
    A[index, j] = 1
A, A.shape, np.linalg.matrix_rank(A)
```

```python
x_t = A[:, np.random.randint(0, n_s)]
#for i in range(n_s): print(np.linalg.norm(x_t - A[:,i]))
```

```python
b = x_t.reshape(D,1)
#print(b)
```

```python
def f(x):
    return np.linalg.norm(A.dot(x) - b)

def make_eq1():
    return lambda x: np.dot(np.ones(x.T.shape), x) - 1

def make_ineq(i, n_s):
    A = np.zeros([n_s, n_s])
    A[i,i] = 1
    return lambda x: np.dot(np.ones([1, n_s]), np.dot(A, x))

def makeConsIneq(n_s):
    f = []
    for i in range(n_s):
        f.append(make_ineq(i, n_s))
    f.append(make_eq1())
    return f
```

```python
cons_funs = makeConsIneq(n_s)

cons = [{"type": "ineq", "fun": func} if i < n_s-1 else {"type": "eq", "fun": func} for i, func in enumerate(cons_funs)]
x = np.array([1 for i in range(n_s)]).reshape([n_s, 1])

result = minimize(f, x0=x, constraints=cons, method="SLSQP")
print(result)
#for i in range(n_s): print(result["x"][i])
print(np.var(result["x"]), np.max(result["x"]), np.min(result["x"]), np.sum(result["x"]))
```

```python
x_opt1 = np.array(result["x"])
np.linalg.norm(A.dot(x_opt1) - b), x_opt1.sum()
```

```python
cons_funs = makeConsIneq(n_s)

cons = [{"type": "ineq", "fun": func} if i < n_s-1 else {"type": "eq", "fun": func} for i, func in enumerate(cons_funs)]
x = np.array([0 for i in range(n_s)]).reshape([n_s, 1])

result = minimize(f, x0=x, constraints=cons, method="SLSQP")
print(result)
#for i in range(n_s): print(result["x"][i])
print(np.var(result["x"]), np.max(result["x"]), np.min(result["x"]), np.sum(result["x"]))
```

```python
x_opt2 = np.array(result["x"])
np.linalg.norm(A.dot(x_opt2) - b), x_opt2.sum()
```

```python
LA = np.zeros([n_s+1+n_s, n_s+1+n_s])
for i in range(n_s+1+n_s):
    if i < n_s:
        for j in range(n_s+1+n_s):
            if j < n_s:
                LA[i,j] = 2*np.dot(A[:,i], A[:,j])
            elif j == n_s:
                LA[i,j] = -1
            elif j > n_s:
                LA[i,j] = 1 if i == (j - (n_s+1)) else 0
                
    elif i == n_s:
        LA[i,0:n_s] = 1
    elif i > n_s:
        LA[i, i-(n_s+1)] = 1
        
LA
```

```python
Lb = np.zeros([(n_s+1+n_s), 1])
for i in range(n_s+1):
    if i < n_s: Lb[i] = 2*np.dot(A[:,i], b)
    elif i == n_s: Lb[i] = 1
Lb
```

```python
Lx = np.linalg.solve(LA, Lb)
print(Lx)
print(Lx[0:n_s].sum())
```

```python
Lb = np.zeros([(n_s+1), 1])
for i in range(n_s+1):
    if i < n_s: Lb[i] = A.T.dot(b)[i,0]
    elif i == n_s: Lb[n_s] = 1
Lb
```

```python
Lx = np.linalg.solve(LA, Lb)
print(Lx, Lx[0:n_s].sum())
print(Lx[Lx<0])
```
